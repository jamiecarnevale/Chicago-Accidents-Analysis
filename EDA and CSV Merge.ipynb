```python
#data from:
# https://data.cityofchicago.org/Transportation/Traffic-Crashes-Crashes/85ca-t3if
# https://data.cityofchicago.org/Transportation/Traffic-Crashes-People/u6pd-qa9d
# https://data.cityofchicago.org/Transportation/Traffic-Crashes-Vehicles/68nd-jvt3

df1 = pd.read_csv("data/Traffic_Crashes_People.csv")
df2 = pd.read_csv("data/Traffic_Crashes_Crashes.csv")
df3 = pd.read_csv("data/Traffic_Crashes_Vehicles.csv")
```

# Merge by Crash Record ID

```python
df1 = df1.drop_duplicates(subset = 'CRASH_RECORD_ID')
df2 = df2.drop_duplicates(subset = 'CRASH_RECORD_ID')
df3 = df3.drop_duplicates(subset = 'CRASH_RECORD_ID')
```

```python
df_merged = df1.merge(df2, how = 'inner',
                     on = 'CRASH_RECORD_ID').merge(df3, how = 'inner',
                                                  on = 'CRASH_RECORD_ID')
```

```python
df_merged.info(verbose=True, null_counts=True)
```

# Exploring and Exporting Merged Dataset
